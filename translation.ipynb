```python
from transformers import pipeline
from langdetect import detect

# translator pipeline

translator = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

# Language Codes

```python
LANG_CODES = {
    "ar": "arb_Arab",  
    "arz": "arz_Arab", 
    "es": "spa_Latn",  
    "de": "deu_Latn",  
    "fr": "fra_Latn",  
    "en": "eng_Latn"   
}
```

# Translation Function

```python
def translate_text(text, target_lang):
    detected_lang = detect(text)
    src_code = LANG_CODES.get(detected_lang, "eng_Latn")
    tgt_code = LANG_CODES[target_lang]
    result = translator(text, src_lang=src_code, tgt_lang=tgt_code)
    return result[0]['translation_text']
```

```python
input_text = """The Palestinian people have the right to live in peace and dignity, 
free from occupation and injustice, and the world must unite to support their cause."""
```

# Target languages

```python
target_languages = ["arz", "es", "de", "fr"]
```

```python
print("(English):")
print(input_text)
print("\n Translation:")
```

# Multi-language Translation

```python
for lang in target_languages:
    translated = translate_text(input_text, lang)
    print(f"\n({lang.upper()}):\n{translated}")
```
